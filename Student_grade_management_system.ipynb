# Student Grade Management System

## Description 

### A simple Python project that allows users to manage student records. Users can add students with their marks, view all student records, find the student with the highest marks (topper), and calculate the average marks of all students. The project uses functions, lists, dictionaries, loops, and conditional statements to demonstrate basic Python programming concepts.

```python
students = []


def add_student():
    name = input("Enter student name: ")
    marks = float(input("Enter marks: "))

    student = {
        "name": name,
        "marks": marks
    }

    students.append(student)
    print("Student added successfully!\n")


def view_students():
    if len(students) == 0:
        print("No students found.\n")
        return

    print("\nStudent Records")
    print("-" * 20)

    for student in students:
        print(f"Name: {student['name']}")
        print(f"Marks: {student['marks']}")
        print("-" * 20)


def find_topper():
    if len(students) == 0:
        print("No students found.\n")
        return

    topper = students[0]

    for student in students:
        if student["marks"] > topper["marks"]:
            topper = student

    print("\nTopper Details")
    print(f"Name: {topper['name']}")
    print(f"Marks: {topper['marks']}\n")


def calculate_average():
    if len(students) == 0:
        print("No students found.\n")
        return

    total = 0

    for student in students:
        total += student["marks"]

    average = total / len(students)

    print(f"\nAverage Marks: {average:.2f}\n")


while True:
    print("===== Student Grade System =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Find Topper")
    print("4. Calculate Average")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        view_students()

    elif choice == "3":
        find_topper()

    elif choice == "4":
        calculate_average()

    elif choice == "5":
        print("Program Closed.")
        break

    else:
        print("Invalid choice. Try again.\n")
```
